```python
import pandas as pd
import numpy as np
# Our diamictite data

Mozaan_Si_Iso = -1.18
Mozaan_Silica = 57.2
Afrikander_Si_Iso = -.25
Afrikander_Silica = 63.3
Promise_Si_Iso = -.34
Promise_Silica = 70.3
Coronation_Si_Iso = -.31
Coronation_Silica = 72
Duitschland_Si_Iso = -0.09
Duitschland_Silica = 55.1
Timeball_Si_Iso = -.38
Timeball_Silica = 65
Makganyene_Si_Iso = -0.69
Makganyene_Silica = 56.7
Ramsay_Si_Iso = -0.22
Ramsay_Silica = 67.9
Bruce_Si_Iso = -.2
Bruce_Silica = 67.9
Gowganda_Si_Iso = -0.05
Gowganda_Silica = 65.8
Wyoming_Si_Iso = -0.06
Wyoming_Silica = 71
Gaskiers_Si_Iso = -.3
Gaskiers_Silica = 66.8
Pocatello_Si_Iso = -.27
Pocatello_Silica = 69.3
Konnarock_Si_Iso = -.26
Konnarock_Silica = 67.5
Nantuo_Si_Iso = -.26
Nantuo_Silica = 65.1
Gucheng_Si_Iso = -0.26
Gucheng_Silica = 66.5
Blaubeker_Si_Iso = -.15
Blaubeker_Silica = 76.8
Kaigas_Si_Iso = -.31
Kaigas_Silica = 63.8
Chuos_Si_Iso = -.31
Chuos_Silica = 47.1
Numees_Si_Iso = -.32
Numees_Silica = 70.7
Ghaub_Si_Iso = -0.26
Ghaub_Silica = 32
Bolivia_Si_Iso = -.28
Bolivia_Silica = 73.6
DwykaE_Si_Iso = -.3
DwykaE_Silica = 67.3
DwykaW_Si_Iso = -.44
DwykaW_Silica = 42.8

# Savage et al., 2014
Basalt_Si_Iso = -0.29
Basalt_Si_iso_sd = 0.07
Basalt_Silica = 50
# Rudnick and Gao, 2014
Avg_UCC_Silica = 66.62
# Savage et al., 2013
Avg_UCC_Si_Iso = -0.25

# South African TTG isotope sources: Andre et al., 2019 and Deng et al., 2019
TTG_Si_Iso = [np.random.normal(loc=-0.01486, scale=0.0566,size= 10000)]
TTG_Si_Iso_sd = np.std(a= TTG_Si_Iso)
TTG_Si_Iso_avg = np.mean(a= TTG_Si_Iso)
# South African TTG silica sources: Andre et al., 2019 and Deng et al., 2019
TTG_Silica = [np.random.normal(loc=71.042162, scale=3.513,size=10000)]
TTG_Silica_sd = np.std(a= TTG_Silica)
TTG_Silica_avg = np.mean(a= TTG_Silica)
# Pongola BIF isotope source: Delvigne et al., 2012
BIF_Si_Iso = [np.random.normal(loc=-1.4499, scale=0.4592, size=10000)]
BIF_Si_Iso_sd = np.std(a= BIF_Si_Iso)
BIF_Si_Iso_avg = np.mean(a= BIF_Si_Iso)
# Pongola BIF silica source: Delvigne et al., 2012
BIF_Silica = [np.random.normal (loc=59.55, scale=5, size=10000)]
BIF_Silica_sd = np.std(a= BIF_Silica)
BIF_Silica_avg = np.mean(a= BIF_Silica)
# Archaean chert isotope sources: Van den Boorn et al., 2007 and 2010, Steinhoefel et al., 2009, Chakrabarti et al., 2012, Delvigne et al., 2012
Chert_Si_Iso = [np.random.normal(loc=-1.051, scale=.086, size=10000)]
Chert_Si_Iso_sd = np.std(a= Chert_Si_Iso)
Chert_Si_Iso_avg = np.mean(a= Chert_Si_Iso)
# Archaean chert silica sources: Van den Boorn et al., 2007 and 2010, Steinhoefel et al., 2009, Chakrabarti et al., 2012, Delvigne et al., 2012
Chert_Silica = [np.random.normal(loc=90.288, scale=5, size=10000)]
Chert_Silica_sd = np.std(a= Chert_Silica)
Chert_Silica_avg = np.mean(a= Chert_Silica)
f_b_values = np.linspace(0, 1, 10000, dtype=float)
f_a_values = np.linspace (1, 0, 10000, dtype=float)

#TTG and Basalt mixing line
TTG_Si_fa = np.multiply(TTG_Silica,f_a_values)
TTG_Si_fa = TTG_Si_fa.T
Basalt_Si_fb = np.multiply(Basalt_Silica,f_b_values)
Basalt_Si_fb = Basalt_Si_fb.T
data = pd.DataFrame(f_a_values)
data['TTG Si fa'] = TTG_Si_fa
data['Bas Si fb']= Basalt_Si_fb
data["TB X mix value"] = data['TTG Si fa'] + data['Bas Si fb']
TTG_Iso_fa = np.multiply(TTG_Si_Iso,f_a_values)
TTG_Iso_fa = TTG_Iso_fa.T
Basalt_Iso_fb = np.multiply(Basalt_Si_Iso,f_b_values)
Basalt_Iso_fb = Basalt_Iso_fb.T
data['TTG Iso fa'] = TTG_Iso_fa
data['Bas Iso fb']= Basalt_Iso_fb
TTG_Si = np.array(TTG_Silica)
TTG_Si = TTG_Si.T
Bas_Silica = np.full((10000), 50, dtype=float)
data['TTG Silica']= TTG_Si
data['Bas Silica']= Bas_Silica
data['TTG Si div X']= data['TTG Silica']/data['TB X mix value']
data['Bas Si div X']= data['Bas Silica']/data['TB X mix value']
data['TB Y mix value']= data['TTG Iso fa']*data['TTG Si div X']+data['Bas Iso fb']*data['Bas Si div X']

# TTG and Chert mixing line
Chert_Si_fb = np.multiply(Chert_Silica,f_b_values)
Chert_Si_fb = Chert_Si_fb.T
TTG_Si_fa = np.multiply(TTG_Silica,f_a_values)
TTG_Si_fa = TTG_Si_fa.T
data2 = pd.DataFrame(f_a_values)
data2['Chert Si fb'] = Chert_Si_fb
data2['TTG Si fa']= TTG_Si_fa
data2["TC X mix value"] = data2['Chert Si fb'] + data2['TTG Si fa']
Chert_Iso_fb = np.multiply(Chert_Si_Iso,f_b_values)
Chert_Iso_fb = Chert_Iso_fb.T
TTG_Iso_fa = np.multiply(TTG_Si_Iso,f_a_values)
TTG_Iso_fa = TTG_Iso_fa.T
data2['Chert Iso fb'] = Chert_Iso_fb
data2['TTG Iso fa']= TTG_Iso_fa
Chert_Si = np.array(Chert_Silica)
Chert_Si = Chert_Si.T
data2['Chert Silica']= Chert_Si
data2['Chert Si div X']= data2['Chert Silica']/data2["TC X mix value"]
data2['TTG Si div X 2']= data['TTG Silica']/data2["TC X mix value"]
data2['TC Y mix value']= data2['Chert Iso fb']*data2['Chert Si div X']+data2['TTG Iso fa']*data2['TTG Si div X 2']

# BIF and Chert mixing line
BIF_Si_fb = np.multiply(BIF_Silica,f_b_values)
BIF_Si_fb = BIF_Si_fb.T
Chert_Si_fa = np.multiply(Chert_Silica,f_a_values)
Chert_Si_fa = Chert_Si_fa.T
data3 = pd.DataFrame(f_a_values)
data3['BIF Si fb'] = BIF_Si_fb
data3['Chert Si fa']= Chert_Si_fa
data3["BC X mix value"] = data3['BIF Si fb'] + data3['Chert Si fa']
BIF_Iso_fb = np.multiply(BIF_Si_Iso,f_b_values)
BIF_Iso_fb = BIF_Iso_fb.T
Chert_Iso_fa = np.multiply(Chert_Si_Iso,f_a_values)
Chert_Iso_fa = Chert_Iso_fa.T
data3['BIF Iso fb'] = BIF_Iso_fb
data3['Chert Iso fa']= Chert_Iso_fa
BIF_Si = np.array(BIF_Silica)
BIF_Si = BIF_Si.T
data3['BIF Silica']= BIF_Si
data3['BIF Si div X']= data3['BIF Silica']/data3["BC X mix value"]
data3['Chert Si div X 2']= data2['Chert Silica']/data3["BC X mix value"]
data3['BC Y mix value']= data3['BIF Iso fb']*data3['BIF Si div X']+data3['Chert Iso fa']*data3['Chert Si div X 2']

# Basalt and BIF mixing line
Basalt_Si_fa = np.multiply(Basalt_Silica,f_a_values)
Basalt_Si_fa = Basalt_Si_fa.T
BIF_Si_fb = np.multiply(BIF_Silica,f_b_values)
BIF_Si_fb = BIF_Si_fb.T
data4 = pd.DataFrame(f_a_values)
data4['Bas Si fa'] = Basalt_Si_fa
data4['BIF Si fb']= BIF_Si_fb
data4["BB X mix value"] = data4['Bas Si fa'] + data4['BIF Si fb']
Bas_Iso_fa = np.multiply(Basalt_Si_Iso,f_a_values)
Bas_Iso_fa = Bas_Iso_fa.T
BIF_Iso_fb = np.multiply(BIF_Si_Iso,f_b_values)
BIF_Iso_fb = BIF_Iso_fb.T
data4['Bas Iso fa'] = Bas_Iso_fa
data4['BIF Iso fb']= BIF_Iso_fb
data4['Bas Si div X']= data['Bas Silica']/data4["BB X mix value"]
data4['BIF Si div X']= data3['BIF Silica']/data4["BB X mix value"]
data4['BB Y mix value']= data4['Bas Iso fa']*data4['Bas Si div X']+data4['BIF Iso fb']*data4['BIF Si div X']

# BIF and TTG mixing line
TTG_Si_fa = np.multiply(TTG_Silica,f_a_values)
TTG_Si_fa = TTG_Si_fa.T
data5 = pd.DataFrame(f_a_values)
data5['TTG Si fa'] = TTG_Si_fa
data5['BIF Si fb']= BIF_Si_fb
data5['BT X mix value'] = data5['TTG Si fa'] + data5['BIF Si fb']
data5['TTG Iso fa'] = TTG_Iso_fa
data5['BIF Iso fb']= BIF_Iso_fb
data5['TTG Si div X']= data['TTG Silica']/data5['BT X mix value']
data5['BIF Si div X']= data3['BIF Silica']/data5['BT X mix value']
data5['BT Y mix value']= data5['TTG Iso fa']*data5['TTG Si div X']+data5['BIF Iso fb']*data5['BIF Si div X']

trendline_data = pd.read_excel('Si mixing trendline.xlsx')

import matplotlib.pyplot as plt
x = ([data['TB X mix value']])
y = ([data['TB Y mix value']])
plt.scatter(x,y, s=.5, c="lightgrey")
x2= (data2['TC X mix value'])
y2= (data2['TC Y mix value'])
plt.scatter(x2,y2, s=.5, c="lightgrey")
x3= (data3['BC X mix value'])
y3= (data3['BC Y mix value'])
plt.scatter(x3,y3, s=.5, c="lightgrey")
x4 = (data4['BB X mix value'])
y4 = (data4['BB Y mix value'])
plt.scatter(x4,y4, s=.5, c="lightgrey")
x5 = (data5['BT X mix value'])
y5 = (data5['BT Y mix value'])
plt.scatter(x5,y5, s=.5, c="lightgrey")
plt.rcParams.update({'font.size': 12})
axes = plt.gca()
axes.set_xlim([40,95])
axes.set_ylim([-1.8, 0.2])
axes.set_xlabel('$SiO_2$(wt.%)')
axes.set_ylabel('$𝛿^{30}Si(‰)$')

plt.rcParams.update({'font.size': 12})

TTG_x = TTG_Silica_avg
TTG_y = TTG_Si_Iso_avg
plt.scatter(TTG_x,TTG_y,s=50, c="black",marker="D")
# plt.text(TTG_x,TTG_y,'Avg S African TTG')
BIF_x = BIF_Silica_avg
BIF_y = BIF_Si_Iso_avg
plt.scatter(BIF_x,BIF_y,s=55, c="black",marker="P")
# plt.text(BIF_x,BIF_y,'Avg Pongola BIF')
Chert_x = Chert_Silica_avg
Chert_y = Chert_Si_Iso_avg
plt.scatter(Chert_x,Chert_y,s=55, c="black",marker="X")
# plt.text(Chert_x,Chert_y,'Avg Archaean Chert')
Basalt_x = Basalt_Silica
Basalt_y = Basalt_Si_Iso
plt.scatter(Basalt_x,Basalt_y,s=50, c="black", marker="s")
Avg_UCC_x= Avg_UCC_Silica
Avg_UCC_y= Avg_UCC_Si_Iso
plt.scatter(Avg_UCC_x,Avg_UCC_y,s=50, c="orange", marker="P")
#plt.text(Avg_UCC_x,Avg_UCC_y, 'Modern UCC')
#plt.text(Basalt_x,Basalt_y,'Avg Basalt')
#plt.plot([Basalt_x,TTG_x],[Basalt_y,TTG_y],c="black", linewidth=.5)
plt.plot([Basalt_x,Avg_UCC_x],[Basalt_y,Avg_UCC_y],c="black", linewidth=1.25, linestyle=':')
#plt.plot([Basalt_x,BIF_x],[Basalt_y,BIF_y],c="black", linewidth=.5)
#plt.plot([Chert_x,TTG_x],[Chert_y,TTG_y],c="black", linewidth=.5)
plt.plot([Chert_x,Avg_UCC_x],[Chert_y,Avg_UCC_y],c="black", linewidth=1.25, linestyle=':')
#plt.plot([Chert_x,BIF_x],[Chert_y,BIF_y],c="black", linewidth=.5)
#plt.plot([BIF_x,TTG_x],[BIF_y,TTG_y],c="black", linewidth=.5)
plt.plot([BIF_x,Avg_UCC_x],[BIF_y,Avg_UCC_y],c="black", linewidth=1.25, linestyle=':')

Sil_Kom_x = 42.5
Sil_Kom_y = -0.29
plt.scatter(Sil_Kom_x, Sil_Kom_y, s=50, marker='s', c='black')
plt.plot([BIF_x,Sil_Kom_x],[BIF_y,Sil_Kom_y],c="black", linewidth=1.25, linestyle=':')
plt.plot([TTG_x,Sil_Kom_x],[TTG_y,Sil_Kom_y],c="black", linewidth=1.25, linestyle=':')

Mozaan_x = Mozaan_Silica
Mozaan_y = Mozaan_Si_Iso
Makganyene_x=Makganyene_Silica
Makganyene_y= Makganyene_Si_Iso
plt.scatter([Mozaan_x,Afrikander_Silica, Promise_Silica,Coronation_Silica],[Mozaan_y, Afrikander_Si_Iso, Promise_Si_Iso, Coronation_Si_Iso],s=50, c="black", edgecolor='black')
plt.scatter([Makganyene_x, Timeball_Silica, Bruce_Silica, Duitschland_Silica, Ramsay_Silica,Gowganda_Silica, Wyoming_Silica],[Makganyene_y, Timeball_Si_Iso,Bruce_Si_Iso, Duitschland_Si_Iso, Ramsay_Si_Iso,Gowganda_Si_Iso, Wyoming_Si_Iso],s=50, c="blue", edgecolor='black')
plt.scatter([Gaskiers_Silica, Pocatello_Silica, Konnarock_Silica, Nantuo_Silica, Blaubeker_Silica, Kaigas_Silica, Numees_Silica, Gucheng_Silica, Chuos_Silica, Ghaub_Silica],[Gaskiers_Si_Iso, Pocatello_Si_Iso, Konnarock_Si_Iso, Nantuo_Si_Iso, Blaubeker_Si_Iso, Kaigas_Si_Iso, Numees_Si_Iso, Gucheng_Si_Iso, Chuos_Si_Iso, Ghaub_Si_Iso],s=50, c="red", edgecolor='black')
plt.scatter([Bolivia_Silica, DwykaE_Silica, DwykaW_Silica],[Bolivia_Si_Iso, DwykaE_Si_Iso, DwykaW_Si_Iso],s=50, c="forestgreen", edgecolor='black')
#plt.text(Mozaan_x,Mozaan_y,'Mozaan')
#plt.text(Makganyene_x,Makganyene_y,'Makganyene')
plt.scatter([trendline_data['TCx'],trendline_data['TBx'],trendline_data['BTx'],trendline_data['BBx'], trendline_data['BCx'], trendline_data['CBx']],[trendline_data['TCy'],trendline_data['TBy'],trendline_data['BTy'],trendline_data['BBy'], trendline_data['BCy'], trendline_data['CBy']], s=0.015, c="black", marker='.')
plt.savefig('Diamictitemixingplot.png', dpi=300, bbox_inches='tight')
plt.show()
```
